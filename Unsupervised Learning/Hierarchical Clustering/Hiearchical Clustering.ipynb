```python
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10, 7))  
plt.title("Dendrograms")  
dend = shc.dendrogram(shc.linkage(data, method='ward'))
plt.axhline(y=190,color='r',linestyle='--')
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=3, metric='euclidean', 
                                  linkage='ward')  
Y_pred=cluster.fit_predict(data)
print(Y_pred)
```

```python
data['clusters']=Y_pred
```

```python
data
```

```python
data['clusters']=data['clusters'].map({0:'Negative',1:'Nuetral',2:'Positive'})
```

```python
data['clusters'].value_counts().plot(kind='pie',autopct='%.2f%%')
```
