```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings("ignore")
```

```python
df= pd.read_csv(r"Mall_Customers.csv",index_col=0,header=0)
df
                      
```

```python
df.describe(include='all')
```

```python
df["Gender"]=df.Gender.map({"Female":0,"Male":1})
```

```python
df["Gender"].value_counts().plot(kind="pie",autopct="%.2f%%");
```

```python
X=  df.values[:,[2,3]]
```

```python
# Using the elbow method to find the optimal number of clusters
from sklearn.cluster import KMeans
 
wsse = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i,  random_state = 10)
    kmeans.fit(X)
    wsse.append(kmeans.inertia_)
    
plt.plot(range(1, 11), wsse,c="blue")
plt.scatter(range(1, 11),wsse,c="red")
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WSSE')
plt.show()
```

```python
print(wsse)
```

```python
# Import the KMeans class from sklearn.cluster
from sklearn.cluster import KMeans

# Fitting K-Means to the dataset
kmeans = KMeans(n_clusters = 5, random_state = 10)
Y_pred = kmeans.fit_predict(X)
 
#kmeans.fit(X)--->training
#Y_pred=kmeans.predict(X)--->predicting
```

```python
Y_pred
```

```python
kmeans.inertia_
```

```python
kmeans.n_iter_
```

```python
df["Clusters"]=Y_pred
df
```

```python
kmeans.cluster_centers_
```

```python
sns.lmplot(data=df, x="Annual Income (k$)",y="Spending Score (1-100)",
          fit_reg=False, #No regression line
          hue="Clusters",palette="Set1")
 
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],
           s=300,c="black")
plt.show()
```

```python
df["Clusters"]=df["Clusters"].map({0:"Standard",1:"Valuable",2:"Most Valuable",3:"Elite",4:"Careful"})
df
```
