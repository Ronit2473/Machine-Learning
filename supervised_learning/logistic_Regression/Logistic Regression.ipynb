```python
X = df.loc[:, df.columns != 'Quit the Company']
y = df.loc[:, 'Quit the Company']

```

```python
X.columns
```

```python
X.shape
y.shape
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X=ss.fit_transform(X)
print(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=19)
```

```python
X_train.shape
y_train.shape
y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
lc=LogisticRegression()

lc.fit(X_train,y_train)
```

```python
Y_pred_proba=lc.predict_proba(X_test)
Y_pred_proba
```

```python
Y_pred=lc.predict(X_test)
Y_pred
```

```python
print(list(zip(y_test,Y_pred)))
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report

cfm=confusion_matrix(y_test,Y_pred)
print(cfm)

print("Classification report: ")

print(classification_report(y_test,Y_pred))

acc=accuracy_score(y_test, Y_pred)
print("Accuracy of the model: ",acc)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

cfm=confusion_matrix(y_test,Y_pred)
sns.heatmap(cfm, annot=True, fmt='g', cbar=False, cmap='BuPu')
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.show()

print("Classification report:")
print(classification_report(y_test,Y_pred))

acc=accuracy_score(y_test,Y_pred)
print("Accuracy of the model: ",acc)
```

```python
y_pred_class=[]
for value in Y_pred_proba[:,1]:
    if value>0.39:
        y_pred_class.append(1)
    else:
        y_pred_class.append(0)
#y_pred_class
```

```python
Y_pred_proba[:,1]
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

cfm=confusion_matrix(y_test,y_pred_class)
sns.heatmap(cfm, annot=True, fmt='g', cbar=False, cmap='BuPu')
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.show()


print("Classification report:")

print(classification_report(y_test,y_pred_class))

acc=accuracy_score(y_test,y_pred_class)
print("Accuracy of the model: ",acc)
```

```python
for i in np.arange(0.40,0.61,0.01):
    predict_mine=np.where(Y_pred_proba[:,1]>i,1,0)
    cfm=confusion_matrix(y_test, predict_mine)
    total_err=cfm[0,1]+cfm[1,0]
    print("Errors at threshold ", i, ":",total_err, " , type 2 error :",
          cfm[1,0]," , type 1 error:", cfm[0,1])
```

```python
classifier=LogisticRegression()

#performing kfold_cross_validation
from sklearn.model_selection import StratifiedKFold
kfold_cv=StratifiedKFold(n_splits=10,random_state=10,shuffle=True)
#print(kfold_cv)

from sklearn.model_selection import cross_val_score
#running the model using scoring metric as accuracy
kfold_cv_result=cross_val_score(estimator=classifier,X=X_train,
                                                 y=y_train, cv=kfold_cv)
print(kfold_cv_result)
#finding the mean
print(kfold_cv_result.mean())
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(classifier, X_train, y_train, cv=5, scoring='accuracy')
print("CV scores:", scores)
print("Mean CV accuracy:", scores.mean())
```
